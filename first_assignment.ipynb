```python
import numpy as np 
import pandas as pd
train_data = {
    'Height (cm)': [160, 170, 175, 180, 158, 165, 172, 178, 162, 177, 159, 168, 173, 181, 161, 176],
    'Weight (kg)': [55, 65, 70, 75, 50, 60, 68, 72, 58, 77, 52, 63, 69, 76, 54, 74],
    'Gender': ['Female', 'Female', 'Male', 'Male', 'Female', 'Female', 'Male', 'Male', 'Female', 'Male', 'Female', 'Female', 'Male', 'Male', 'Female', 'Male']
}
train_delta=pd.DataFrame(train_data)
test_data = {
    'Height (cm)': [163, 179, 167, 174, 155, 182, 160, 171],
    'Weight (kg)': [57, 78, 62, 71, 48, 80, 53, 67],
    'Gender': ['Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male']
}
test_delta = pd.DataFrame(test_data)
# print("train_data")
# print(train_delta)
# print("test_data")
# print(test_delta)
#layer 1 -inputs(2) and layer2-(2) nodes and layer3-(2) output
weights_1=np.array([[-0.5,-1.1],[2.1,-3.5]])
baises_1=np.array([0.1,-0.5])
weights_2=np.array([[-1.5,-0.1],[2.2,-1.5]])
baises_2=np.array([0.1,-0.5])
def relu(x):
    return np.where(x>0,x,0)
def dels(inputs,expect,layer1,layer2,w2):
    del_biase_2=np.array([np.power(-1,expect)*layer2[expect],np.power(-1,expect+1)*layer2[expect]])
    del_biase_1=rlu_der(layer1)*(w2[expect]-w2[1-expect])*layer2[expect]
    del_weights_1=del_biase_1*rlu_der(np.array(np.transpose([layer1,layer1])))*np.array([inputs,inputs])
    a1=layer2[0]
    a2=layer2[1]
    a3=layer1[0]
    a4=layer1[1]
    if(expect==1):
        del_w2=np.array([[-a2*a3,-a2*a4],[a2*a3,a2*a4]])
    if(expect==0):
        del_w2=np.array([[a1*a3,a1*a4],[-a1*a3,-a1*a4]])
        # del_w2=np.array(([layer2[0]*layer1[0],layer2[0]]*layer1[1],[-layer2[0]*layer1[0],-layer2[0]*layer1[1]]))
    return del_weights_1,del_w2,del_biase_1,del_biase_2
def rlu_der(x):
    return np.where(x>0,1,0)
def layer1(weights1,biases1,inputs):
    inputs=inputs.copy()
    biases1=biases1.copy()
    weights1=weights1.copy()
    input_array=inputs.reshape((2,1))
    baises=biases1.reshape((2,1))
    layer1=np.dot(weights1,input_array)+baises
    layer1=layer1.reshape(2)
    return relu(layer1)
def dange_layer1(weights1,biases1,inputs):
    inputs=inputs.copy()
    biases1=biases1.copy()
    weights1=weights1.copy()
    input_array=inputs.reshape((2,1))
    baises=biases1.reshape((2,1))
    layer1=np.dot(weights1,input_array)+baises
    layer1=layer1.reshape(2)
    return layer1
def layer2(layer1,biases2,weights2):
    layer1=layer1.copy()
    biases2=biases2.copy()
    weights2=weights2.copy()
    input=layer1.reshape((2,1))
    biases=biases2.reshape((2,1))
    layermain_2=np.dot(weights2,input)+biases
    layermain_2=np.array([np.exp(layermain_2[0][0])/(np.exp(layermain_2[0][0])+np.exp(layermain_2[1][0])),np.exp(layermain_2[1][0])/(np.exp(layermain_2[0][0])+np.exp(layermain_2[1][0]))])
    return layermain_2

def Error_function(expect,weights1,weights2,biases1,biases2,inputs):
    layer_12=layer1(weights1,biases1,inputs)
    layer_22=layer2(layer_12,biases2,weights2)
    return -np.log10(layer_22[1-expect])
def update(w1,w2,b1,b2):
    new_weights_1=w1-np.sum(train_delta['delw1'],axis=0)*0.25
    new_weights_2=w2-np.sum(train_delta['delw2'],axis=0)*0.25
    new_biases_1=b1-np.sum(train_delta['delb1'],axis=0)*0.25
    new_biases_2=b2-np.sum(train_delta['delb2'],axis=0)*0.25
    return new_weights_1,new_weights_2,new_biases_1,new_biases_2
def net_error(data_frame, weights1, weights2, biases1, biases2):
    error = 0
    error_frame = []
    a1=[]
    a2=[]
    a3=[]
    a4=[]
    for i in range(len(data_frame)):
        inputs = np.array([data_frame.iloc[i, 0], data_frame.iloc[i, 1]])
        # print(inputs)
        expect = 1 if data_frame.iloc[i, 2] == "Male" else 0
        error_value = Error_function(expect, weights1, weights2, biases1, biases2, inputs)
        error_frame.append(error_value)
        error += error_value
        x1,x2,x3,x4=dels(inputs,expect,layer1(weights1,biases1,inputs),layer2(layer1(weights1,biases1,inputs),biases2,weights2),weights2)
        # x1=layer1(weights1,biases1,inputs)
        # print(x1)
        # x2=layer2(layer1(weights1,biases1,inputs),biases2,weights2)
        # print(x2)
        a1.append(x1)
        a2.append(x2)
        a3.append(x3)
        a4.append(x4)
    data_frame['error'] = error_frame
    data_frame['delw1'] =a1
    data_frame['delw2']=a2
    data_frame['delb1']=a3
    data_frame['delb2']=a4
    return error
print(net_error(train_delta,weights_1,weights_2,baises_1,baises_2))
new_weights_1=weights_1-np.sum(train_delta['delw1'],axis=0)*0.25
new_weights_2=weights_2-np.sum(train_delta['delw2'],axis=0)*0.25
new_biases_1=baises_1-np.sum(train_delta['delb1'],axis=0)*0.25
new_biases_2=baises_2-np.sum(train_delta['delb2'],axis=0)*0.25
print(net_error(train_delta,new_weights_1,new_weights_2,new_biases_1,new_biases_2))
w1,w2,b1,b2=update(new_weights_1,new_weights_2,new_biases_1,new_biases_2)
for i in range(10):
    print(net_error(train_delta,w1,w2,b1,b2))
    w1,w2,b1,b2=update(w1,w2,b1,b2)
print("w1 :",w1)
print("\n")
print("w2 :  ",w2)
print("\n")
print("b1 :",b1)
print("b2 :",b2)
print("minimum_error_taken",net_error(train_delta,w1,w2,b1,b2))
z=[]
for i in range(len(test_delta)):
     inputs = np.array([test_delta.iloc[i, 0],test_delta.iloc[i, 1]])
     print(inputs)
     output_layer_1=layer1(w1,b1,inputs)
     print(output_layer_1)
     print(dange_layer1(w1,b1,inputs))
     output_layer_2=layer2(output_layer_1,b2,w2)
     print(output_layer_2)
     x1="Male" if(output_layer_2[1]>output_layer_2[0]) else "Female"
     print(x1)
     z.append(x1)
test_delta['output']=z
print(test_delta)


# print(train_delta)
```
